複数のグラフを描く

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 2*np.pi)
y1 = np.sin(x)
y2 = np.cos(x)


positions = [0, np.pi/2, np.pi, np.pi*3/2, np.pi*2]
labels = ["0°", "90°", "180°", "270°", "360°"]

plt.title("graphs of trigonometric functions")
# グラフのx軸とy軸に名前を設定します
plt.xlabel("x-axis")
plt.ylabel("y-axis")
# グラフにグリッドを表示します
plt.grid(True)
# グラフのx軸にラベルを設定します
plt.xticks(positions, labels)
# データx, y1をグラフにプロットし、黒で表示してください
plt.plot(x, y1, color="k") 
# データx, y2をグラフにプロットし、青で表示してください
plt.plot(x, y2, color="b") 
plt.show()
```

複数のグラフを書く2

```python
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(2, 3, 1)
ax.plot(x,y1)
ax = fig.add_subplot(2, 3, 2)
ax.plot(x,y1)
ax = fig.add_subplot(2, 3, 3)
ax.plot(x,y2)
ax = fig.add_subplot(2, 3, 4)
ax.plot(x,y2)
ax = fig.add_subplot(2, 3, 5)
ax.plot(x,y2)
ax = fig.add_subplot(2, 3, 6)
ax.plot(x,y2)

plt.show()
```

★plt.subplots_adjust(wspace=1, hspace=1) 
グラフの間隔をあける

```python
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(2, 3, 1)
ax.plot(x,y1)
ax = fig.add_subplot(2, 3, 2)
ax.plot(x,y1)
ax = fig.add_subplot(2, 3, 3)
ax.plot(x,y2)
ax = fig.add_subplot(2, 3, 4)
ax.plot(x,y2)
ax = fig.add_subplot(2, 3, 5)
ax.plot(x,y2)
ax = fig.add_subplot(2, 3, 6)
ax.plot(x,y2)
plt.subplots_adjust(wspace=1, hspace=1)
#plt.subplots_adjust((wspace=2, hspace=2)

plt.show()




```

gridをつける（一番目と６番目に）
一番目にだけｘ、ｙにラベルを付ける

```python
fig = plt.figure(figsize=(9, 6))
#1
ax = fig.add_subplot(2, 3, 1)
ax.grid(True)
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")


ax.plot(x,y1)
#2
ax = fig.add_subplot(2, 3, 2)
ax.plot(x,y1)
#3
ax = fig.add_subplot(2, 3, 3)
ax.plot(x,y2)
#4
ax = fig.add_subplot(2, 3, 4)
ax.plot(x,y2)
#5
ax = fig.add_subplot(2, 3, 5)
ax.plot(x,y2)
#6
ax = fig.add_subplot(2, 3, 6)
ax.grid(True)
ax.plot(x,y2)
plt.subplots_adjust(wspace=1, hspace=1)
#plt.subplots_adjust((wspace=2, hspace=2)

plt.show()
```

サブプロット内のグラフに目盛りを設定する
1番目だけに適用

```python
fig = plt.figure(figsize=(9, 6))
#1
ax = fig.add_subplot(2, 3, 1)
ax.grid(True)
ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel")
positions = [0, np.pi/2, np.pi, np.pi*3/2, np.pi*2]
labels = ["0°", "90°", "180°", "270°", "360°"]


ax.set_xticks(positions)
ax.set_xticklabels(labels)


ax.plot(x,y1)
#2
ax = fig.add_subplot(2, 3, 2)
ax.plot(x,y1)
#3
ax = fig.add_subplot(2, 3, 3)
ax.plot(x,y2)
#4
ax = fig.add_subplot(2, 3, 4)
ax.plot(x,y2)
#5
ax = fig.add_subplot(2, 3, 5)
ax.plot(x,y2)
#6
ax = fig.add_subplot(2, 3, 6)
ax.grid(True)
ax.plot(x,y2)
plt.subplots_adjust(wspace=1, hspace=1)
#plt.subplots_adjust((wspace=2, hspace=2)
```
